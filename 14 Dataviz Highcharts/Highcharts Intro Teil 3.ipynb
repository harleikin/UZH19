```python
from pygments import highlight
from pygments.lexers import PythonLexer
from pygments.lexers import HtmlLexer
from pygments.formatters import HtmlFormatter
import IPython
from IPython.display import HTML
from IPython.display import IFrame

def show_html_file(filename):
    with open(filename) as f:
        code = f.read()
    formatter = HtmlFormatter()
    iframe = IFrame(filename, width=700, height=400)
    code = HTML('<style type="text/css">{}</style>{}'.format(
        formatter.get_style_defs('.highlight'),
        highlight(code, HtmlLexer(), formatter)))
    display(iframe,code)
```

# Zeitreihen

```python
show_html_file("examples/highcharts-date-formats.html")
```

# Drill-Down

```python
show_html_file("examples/highcharts-drilldown.html")
```

# Exportieren

```python
show_html_file("examples/highcharts-exporting.html")
```

```python
show_html_file("examples/highcharts-master-details-preview.html")
```
